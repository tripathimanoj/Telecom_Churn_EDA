# UNIVARIATE ANALYSIS

```python
'''What is the distribution of customer churn in the dataset?

Plot a countplot for the 'Churn' column.'''

# Plot a countplot for the 'Churn' column
sns.countplot(x='Churn', data=data)
plt.title('Distribution of Customer Churn')
plt.show()

# Calculate the total percentage of churn and non-churn customers
churn_percentage = data['Churn'].value_counts(normalize=True) * 100
print(churn_percentage)

print(f'Percentage of Churn Customers: {churn_percentage[True]:.2f}%')
print(f'Percentage of Non-Churn Customers: {churn_percentage[False]:.2f}%')
```

observation: This plot will show the number of customers who churned (True) and those who did not (False).

```python
'''How are international plans distributed among customers?

Plot a countplot for the 'International plan' column.'''

# Plot a countplot for the 'International plan' column
sns.countplot(x='International plan', data=data)
plt.title('Distribution of International Plans')
plt.show()
```

observation: This plot will display the distribution of customers with and without international plans.

```python
'''What percentage of customers have a voice mail plan?
Calculate and visualize the percentage of customers with a voice mail plan.'''

sns.countplot(x='Voice mail plan',data=data)
plt.title("Coustomers vs voice mail plan")
plt.show()
```

```python
# also calculating the percnetage
voice_mail_percnt = data['Voice mail plan'].value_counts(normalize=True) * 100
voice_mail_percnt.plot(kind='bar', rot=0)
plt.title('Percentage of Customers with Voice Mail Plan')
plt.ylabel('Percentage')
plt.show()
```

observation: This bar plot will show the percentage of customers with and without a voice mail plan, here we can see that less customers have registered themself for voice mail plan

```python
'''What is the distribution of the number of customer service calls?
Plot a histogram for the 'Customer service calls' column.'''

# Plot a histogram for the 'Customer service calls' column
plt.hist(data['Customer service calls'], bins=10, edgecolor='black')
plt.title('Distribution of Customer Service Calls')
plt.xlabel('Number of Customer Service Calls')
plt.ylabel('Frequency')
plt.show()
```

Observation: This histogram will give an overview of the distribution of the number of customer service calls.

```python
'''How are account lengths distributed among customers?
Plot a histogram for the 'Account length' column.'''
# Plot a histogram for the 'Account length' column

plt.hist(data['Account length'], bins=20, edgecolor='black')
plt.title('Distribution of Account Lengths')
plt.xlabel('Account Length')
plt.ylabel('Frequency')
plt.show()
```

observation:

Bivariate Analysis

```python
'''What is the relationship between 'Churn' and 'Total day minutes'?
Plot a boxplot or violin plot to compare the distribution of total day minutes for churned and non-churned customers.'''

# Plot a boxplot for 'Total day minutes' with respect to 'Churn'
sns.boxplot(x='Churn', y='Total day minutes', data=data)
plt.title('Relationship between Churn and Total Day Minutes')
plt.show()
print('--------------')

churn_percentage = data[data['Churn'] == True]['Total day minutes'].count() / data['Total day minutes'].count() * 100
non_churn_percentage = 100 - churn_percentage

print(f'Percentage of Churn Customers: {churn_percentage:.2f}%')
print(f'Percentage of Non-Churn Customers: {non_churn_percentage:.2f}%')
```

Observation: This plot will help identify if there's a significant difference in total day minutes between churned and non-churned customers.

```python
'''Are customers with international plans more likely to churn?
Plot a countplot of 'Churn' with hue='International plan'.'''

# Plot a countplot of 'Churn' with hue='International plan'
sns.countplot(x='Churn', hue='International plan', data=data)
plt.title('Churn Distribution by International Plan')
plt.show()
print('---------------------')

churn_percentage = data[data['Churn'] == True]['International plan'].count() / data['International plan'].count() * 100
non_churn_percentage = 100 - churn_percentage

print(f'Percentage of Churn Customers: {churn_percentage:.2f}%')
print(f'Percentage of Non-Churn Customers: {non_churn_percentage:.2f}%')
```

observation: This plot will show the distribution of churn for customers with and without international plans.
So, their are less amount of customers churning who are having or not having international plans

```python
'''How does the distribution of evening call minutes vary with churn?
Plot a boxplot or violin plot for 'Total eve minutes' with respect to 'Churn'.'''

# Plot a boxplot for 'Total eve minutes' with respect to 'Churn'
sns.boxplot(x='Churn', y='Total eve minutes', data=data)
plt.title('Relationship between Churn and Total Eve Minutes')
plt.show()
# Calculate the percentage of churn and non-churn customers
churn_percentage = data[data['Churn'] == True]['Total eve minutes'].count() / data['Total eve minutes'].count() * 100
non_churn_percentage = 100 - churn_percentage

print(f'Percentage of Churn Customers: {churn_percentage:.2f}%')
print(f'Percentage of Non-Churn Customers: {non_churn_percentage:.2f}%')
```

observation: This plot will help identify if there's a significant difference in total evening minutes between churned and non-churned customers.
so, their are less customers who are churning accourding to this service (Total eve minutes)

```python
'''Is there a correlation between the number of voice mail messages and customer churn?
Explore the relationship between 'Number vmail messages' and 'Churn'.'''

# Plot a countplot of 'Churn' with hue='Voice mail plan'
sns.countplot(x='Churn', hue='Voice mail plan', data=data)
plt.title('Churn Distribution by Voice Mail Plan')
plt.show()
```

observation: This plot will show the distribution of churn for customers with and without a voice mail plan.
The number of customer who are having voice mail plan they stuck with company and who are not having plan they also stuck(higher no of customers not having voice mail plan but they stuck with it),
so, their are some amount of customers who left the company those are not having voice mail plan.

```python
'''Do customers who make more customer service calls have a higher likelihood of churning?
Investigate the relationship between 'Customer service calls' and 'Churn'.'''

# Plot a boxplot for 'Customer service calls' with respect to 'Churn'
sns.boxplot(x='Churn', y='Customer service calls', data=data)
plt.title('Relationship between Churn and Customer Service Calls')
plt.show()
```

observation: This plot will help identify if there's a significant difference in the number of customer service calls between churned and non-churned customers.
In this plot we can see that, their are some customers who take service calls multiple time but did't leave it(may be they have no option or old customers), but we can also see that the customers who have taken more than 3 service calles they left the service.
